```python
df = pd.read_excel("/kaggle/input/samsung-healh-daily-summary/com.samsung.shealth.tracker.pedometer_day_summary.20230307161222.xlsx")
df.head()
```

```python
df['create_time'] = pd.to_datetime(df['create_time']) #convert to datetime
df['Year'] = df['create_time'].dt.year.astype('string')
df['Month'] = df['create_time'].dt.month.astype('string')
df['Day'] = df['create_time'].dt.day.astype('string')
df['DayofWeek'] = df['create_time'].dt.dayofweek.astype('string')
df['DayName'] = df['create_time'].dt.day_name().astype('string')

df = df.sort_values('create_time', ascending = True) #sort by date ascending
df.head()
```

```python
df.describe()
```

```python
sns.set_theme()
sns.barplot(x = 'Year', y = 'step_count', data = df)
plt.title('Yearly summary of step count')
plt.show()
```

```python
# Create mappings and replace
mappings = {'Monday':'weekday', 'Tuesday':'weekday', 'Wednesday': 'weekday', 
            'Thursday': 'weekday', 'Friday': 'weekday', 
            'Saturday': 'weekend', 'Sunday': 'weekend'}

df['Day_week'] = df['DayName'].replace(mappings)

sns.lineplot(x = 'Year', y = 'step_count', data = df, hue = 'Day_week')
plt.title('Yearly summary of step count by day of week')
plt.show()
```

```python
sns.histplot(data=df, x="calorie", hue = 'DayName')
plt.title('Calorie distribution over the years')
plt.show()
```

```python
sns.catplot(data=df, x="Year", y="calorie")
plt.xticks(rotation=45)
plt.show()
```

```python
Highest_calories_burned = df[df['calorie'] >= 1000]

print(Highest_calories_burned[['step_count','active_time','create_time','speed','distance','calorie','Year','Month','Day','DayName','Day_week']])
```

```python
df_1 = df[['step_count', 'speed', 'distance','calorie', 'DayName', 'Year']]

sns.jointplot(data = df_1, x = 'distance', y = 'step_count' , hue="DayName", kind = 'scatter')
plt.show()
```

```python
sns.jointplot(data = df_1, x = 'speed', y = 'step_count', kind = 'scatter', hue = 'Year')
plt.show()
```

```python
df_2 = df_1[['step_count', 'speed', 'distance','calorie']]

corr = df_2.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()
```
